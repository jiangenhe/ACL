```python
import pandas as pd
```

```python
df = pd.read_parquet("acl-publication-info.74k.parquet")
```

```python
df.shape
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
# Count the number of articles per year
year_counts = df['year'].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(year_counts.index, year_counts.values)

# Customize the chart
plt.title('Number of Articles by Year')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
# Count the number of articles per booktitle and get top 20
booktitle_counts = df['booktitle'].value_counts().head(20)

# Create a bar chart
plt.figure(figsize=(15, 8))
plt.bar(range(len(booktitle_counts)), booktitle_counts.values)

# Customize the chart
plt.title('Top 20 Publication Venues by Number of Articles')
plt.xlabel('Publication Venue')
plt.ylabel('Number of Articles')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add booktitle labels
plt.xticks(range(len(booktitle_counts)), booktitle_counts.index, rotation=45, ha='right')

# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.show()
```
